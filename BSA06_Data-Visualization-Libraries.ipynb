```python
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
```

```python
tips
```

```python
# 나란히 비교하는 산점도
sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    col="time",
    hue="smoker",
    style="smoker",
    size="size"
)
plt.show()
```

```python
fmri = sns.load_dataset("fmri")
fmri
```

```python
# 통계적 추정
sns.relplot(
    data=fmri,
    kind="line",
    x="timepoint",
    y="signal",
    col="region",
    hue="event",
    style="event",
)
plt.show()
```
