```python
from pyspark.sql import SparkSession

spark =  SparkSession.builder.appName("wine").getOrCreate()
white  = spark.read.csv("white.csv",inferSchema=True,header=True)
```

```python
white.show()
# - 덩어리마다 리스트로 만들어줘야 함 
```

```python
from pyspark.ml.feature import VectorAssembler

설명변수 = list(white.columns) # 일단 컬럼 값을 리스트로
설명변수 = 설명변수[0:-1]
설명변수
```

```python
변수묶음 = VectorAssembler(inputCols=설명변수,outputCol="features") # 레코드 한 줄을 묶음 
변환자료  = 변수묶음.transform(white) # 어셈블러 실행.저장
변환자료.show()
```

```python
변환자료.select("features").show()
```

```python
회귀자료 = 변환자료.select(["features","y"])
```

```python
from pyspark.ml.regression import LinearRegression

train_data, test_data = 회귀자료.randomSplit([0.7, 0.3], 316)
```

```python
회귀분석  = LinearRegression(featuresCol="features",labelCol="y").fit(train_data) # 설명변수:feature, y: label fit: beta를 추정한다 
```

```python
회귀분석.coefficients
```

```python
회귀분석.intercept
```

```python
pred = 회귀분석.evaluate(test_data)
```

```python
pred.predictions.show()
```

```python
# 결정계수, MSE
print(pred.r2, pred.meanSquaredError)
```
