결측값을 통계값으로 대체

```python
from pyspark.ml.feature import Imputer
```

```python
대체 = Imputer(
    inputCols=['educ','salary'],
    outputCols=["{}_imputed".format(c) for c in ['educ','salary']]
).setStrategy('mean')
```

```python
대체.fit(스파크DF).transform(스파크DF).show()
```

```python
대체 = Imputer(
    inputCols=['educ','salary'],
    outputCols=["{}_imputed".format(c) for c in ['educ','salary']]
).setStrategy('mode')

대체.fit(스파크DF).transform(스파크DF).show()
```
