```python
# 카드자료 = pd.read_csv('/Bigdata/archive/creditcard.csv')
카드자료 = pd.read_csv('train/creditcard.csv')
카드자료.info()
```

```python
카드자료["Class"] = 카드자료["Class"].astype('category')
for 변수명 in list(카드자료.columns[:-1]):
    카드자료[변수명] = 카드자료[변수명].astype('float32')
카드자료.info()
```

```python
카드자료.to_parquet('creditcard.parquet')
```

```python
카드자료["Class"].value_counts()
```

```python
X = 카드자료.iloc[:,0:-1] # 모든 행, 모든 열 
X.info()
```

```python
y = 카드자료["Class"]
y.info()
```

```python
from imblearn.over_sampling import SMOTE

X_o, y_o = SMOTE(random_state=316, sampling_strategy=0.5).fit_resample(X,y)
```

```python
카드오버 = pd.concat([X_o, y_o], axis=1)
```

```python
카드오버["Class"].value_counts()
```

```python
from imblearn.under_sampling import NearMiss

X_u,y_u = NearMiss(version=1).fit_resample(X,y)
```

```python
카드언더 = pd.concat([X_u, y_u], axis=1)
```

```python
카드언더["Class"].value_counts()
```

```python
from imblearn.under_sampling import RandomUnderSampler

X_ru, y_ru = RandomUnderSampler(random_state=316, sampling_strategy=0.5).fit_resample(X, y)
```

```python
카드랜덤언더 = pd.concat([X_ru, y_ru], axis=1)
```

```python
카드랜덤언더["Class"].value_counts()
```
