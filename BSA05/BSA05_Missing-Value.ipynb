```python
import pandas as pd
import numpy as np
```

```python
# 결측자료=pd.read_csv("/content/drive/MyDrive/BSA/Employee_missing.csv", encoding="cp949")  # colab
결측자료=pd.read_csv("Employee_missing.csv", encoding="cp949")
```

```python
결측자료.head()
```

```python
결측자료.isna().sum()   # 결측자료 수
```

```python
결측자료.dropna(axis=0).head()  # 결측자료가 하나라도 있는 경우
```

```python
수치변수명=결측자료.select_dtypes(include=np.number).columns.tolist()
```

```python
수치결측자료=결측자료[수치변수명]
```

```python
수치결측자료.head()
```

```python
from sklearn.impute import KNNImputer
```

```python
근접대체=KNNImputer(n_neighbors=5,weights='uniform')
```

```python
근접대체.set_output(transform='pandas')
```

```python
근접대체.fit_transform(수치결측자료).head()
```

```python
수치결측자료.groupby("educ").count()
```

```python
수치결측자료.groupby("educ").mean()
```
