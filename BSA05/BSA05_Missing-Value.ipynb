```python
결측자료=pd.read_csv("/content/drive/MyDrive/BSA/BSA05/Employee_missing.csv", encoding="cp949")  # colab
# 결측자료=pd.read_csv("Employee_missing.csv", encoding="cp949")
결측자료.head()
```

```python
결측자료.isna().sum()   # 결측자료 수
```

```python
# 행 자료 제거

# 결측자료.dropna(axis=0).head() # 결측값이 있음
# 결측자료.dropna(how='all').head()  # 모든 자료가 결측 
k=5
결측자료.dropna(thresh=k, inplace=True) # 결측값이 아닌 개수가 k개 미만
결측자료.dropna(subset=["gender", "jobcat"], inplace=True) # 결측값이 특정 변수에
# 결측자료.dropna(subset="gender", inplace=True) # 결측을 제거하고 동일명의 데이터프레임으로 저장
결측자료.head()
```

```python
# 열 변수 제거

# 결측자료.dropna(axis=1).head()
# 결측자료.dropna(axis='columns').head()
결측자료.dropna(axis=1, thresh=474).head()
결측자료["jobcat"].replace({"경영자":np.NaN, "사무직":np.NaN})
```

결측값 대체

```python
# 직접 지정한 값으로 대체

# 결측자료.fillna(0).head()
결측자료.fillna({"id":0, "gender":"남자", "jobcat":"경영자"}).head()
# 결측자료.fillna(method='ffill').head()
# 결측자료.fillna(method='bfill').head()
```

```python
# 특정 통계값으로 대체 

수치변수명=결측자료.select_dtypes(include=np.number).columns.tolist()
수치결측자료=결측자료[수치변수명]
수치결측자료.head()
```

```python
수치결측자료.groupby("educ").count()
```

```python
수치결측자료.groupby("educ").mean()
```

```python
수치결측자료.fillna(수치결측자료.mean()).head() # median(), min(), max()
```

```python
최빈값 = 결측자료.mode()
최빈값
```

```python
결측자료.fillna(최빈값.iloc[0,:]).head() # integer location[행 인덱스, 열 인덱스] 
```

대체 패키지

```python
# 수치자료의 결측값을 단순대체 
from sklearn.impute import SimpleImputer

단순대체 = SimpleImputer(strategy='mean')
대체자료 = 단순대체.fit_transform(수치결측자료)
대체자료
```

```python
대체자료 = pd.DataFrame(대체자료, columns=수치결측자료.columns)
대체자료.head()
```

```python
단순대체.set_output(transform='pandas')
대체자료 = 단순대체.fit_transform(수치결측자료)
대체자료.head()
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

확률대체 = IterativeImputer(random_state=0)
확률대체.set_output(transform='pandas')
대체자료 = 확률대체.fit_transform(수치결측자료)
대체자료.head()
```

```python
from sklearn.impute import KNNImputer

근접대체=KNNImputer(n_neighbors=5,weights='uniform')
근접대체.set_output(transform='pandas')
근접대체.fit_transform(수치결측자료).head()
```
