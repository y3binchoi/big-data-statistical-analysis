```python
import pandas as pd 
```

```python
df1=pd.read_csv("./combinedata/data1.csv")
df2=pd.read_csv("./combinedata/data2.csv")
df3=pd.read_csv("./combinedata/data3.csv")
display(df1)
```

```python
display(df2)
```

```python
display(df3)
```

Pandas DataFrame
- concat
- merge

```python
df = pd.concat([df1,df2])  # 위아래 붙이기
display(df)
```

```python
df = pd.concat([df1,df2], ignore_index=True)  # 인덱스 새로 편집
display(df)
```

```python
df = pd.concat([df1,df2], keys=(["데이터1","데이터2"]))  # DF에 key값 추기
display(df)
```

```python
df4 = pd.concat([df1,df2], keys=(["데이터1","데이터2"])).drop_duplicates()  # 중복 삭제
# df4.reset_index(inplace=True)
df4.reset_index(drop=True, inplace=True)
display(df4)
```

```python
df = pd.concat([df1,df3])
display(df)
```

```python
df = pd.concat([df1,df3],join='inner')
display(df)
```

```python
df = df1.merge(df2, sort=True)
display(df)
```

```python
df=pd.merge(df1,df2)
display(df)
```

```python
df=pd.merge(df4,df3)
display(df)
```

```python
df=pd.merge(df1,df2, on=['학번'], how='inner')

display(df)  # 특정 변수값을 기준으로 좌우결합
# 각 df가 동일한 변수명을 갖고 있으면 suffixes(_무엇)이 붙음
```

```python
df5 = df2.copy()
df5.rename(columns={'학번':"ID"},inplace=True)
df = pd.merge(df1,df5,left_on='학번',right_on="ID",how='right')
display(df)  # 변수명이 달라도 옆에 붙이기 가능
```

```python
df = pd.merge(df1,df2,on=['이름'],how='inner')
display(df)  # 비교 변수명에 동일한 값이 여러개 있으면 제대로 옆에 붙지 않겠지
```
